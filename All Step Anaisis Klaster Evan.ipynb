>## **MENAMPILKAN GRAFIK ELBOW,REDUKSI STANDARISASI MIN MAX DALAM DATA SET SELURUH VARIABEL (12) DAN JUMLAH DATA (28899)**

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Standarisasi data (jika belum dilakukan)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(numerical_data_cleaned.drop(columns=['Cluster'], errors='ignore'))

# Reduksi dimensi dengan PCA
pca = PCA(n_components=2)  # Misalnya, memilih 2 komponen utama
data_pca = pca.fit_transform(data_scaled)

# Menentukan range jumlah cluster yang ingin diuji
wcss = []  # List untuk menyimpan WCSS untuk setiap jumlah cluster

# Mencari WCSS untuk berbagai jumlah cluster
for i in range(1, 11):  # Menguji 1 hingga 10 cluster
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(data_pca)  # Melakukan clustering pada data yang sudah direduksi
    wcss.append(kmeans.inertia_)

# Plotting grafik elbow untuk menentukan jumlah cluster optimal
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', color='b')
plt.title('Metode Elbow untuk Menentukan Jumlah Cluster')
plt.xlabel('Jumlah Cluster')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()

# Berdasarkan grafik elbow, tentukan jumlah cluster optimal
optimal_clusters = 4  # Ganti dengan jumlah cluster optimal berdasarkan grafik elbow

# Melakukan K-Means clustering dengan jumlah cluster optimal
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
cluster_labels = kmeans.fit_predict(data_pca)

# Menambahkan hasil clustering ke data
numerical_data_cleaned['Cluster'] = cluster_labels

# Menampilkan hasil clustering
print(numerical_data_cleaned.head())
```
